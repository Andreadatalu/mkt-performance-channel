```python
# chat_started

import datetime
import calendar

def leap_day_counter(year):
    leap_days = 0
    # Since 1582, 11 days are missing
    if year >= 1582:
        leap_days += 11
    for specific_year in range(1, year):
        if calendar.isleap(specific_year):
            leap_days += 1
    return leap_days

def month_to_day(year, month):
    all_days = 0
    for specific_month in range(1, month + 1):
        days_in_month = calendar.monthrange(year, specific_month)
        all_days += days_in_month[1]
    return all_days

reference_date = datetime.datetime(2022, 1, 1)  # Reference date (January 1, 2022)

# Create a new list to store the elapsed days
chat_started_days = []

# Iterate over the dates in data['chat_started']
for date_str in data['chat_started']:
    date = datetime.datetime.strptime(date_str, '%d/%m/%Y')  # Convert the string to a datetime object
    days_passed = (
        (date.year - reference_date.year) * 365
        + leap_day_counter(date.year)
        - leap_day_counter(reference_date.year)
        + month_to_day(date.year, date.month)
        - month_to_day(reference_date.year, reference_date.month)
        + date.day
        - reference_date.day
    )
    chat_started_days.append(days_passed)

# Add the list of elapsed days as a new column in data
data['chat_started_days'] = chat_started_days

```

```python
#chat_ended

import datetime
import calendar

def leap_day_counter(year):
    leap_days = 0
    # Since 1582, 11 days are missing
    if year >= 1582:
        leap_days += 11
    for specific_year in range(1, year):
        if calendar.isleap(specific_year):
            leap_days += 1
    return leap_days

def month_to_day(year, month):
    all_days = 0
    for specific_month in range(1, month + 1):
        days_in_month = calendar.monthrange(year, specific_month)
        all_days += days_in_month[1]
    return all_days

reference_date = datetime.datetime(2022, 1, 1)  # Reference date (January 1, 2022)

# Create a new list to store the elapsed days
chat_ended_days = []

# Iterate over the dates in data['chat_started']
for date_str in data['chat_ended']:
    date = datetime.datetime.strptime(date_str, '%d/%m/%Y')  # Convert the string to a datetime object
    days_passed = (
        (date.year - reference_date.year) * 365
        + leap_day_counter(date.year)
        - leap_day_counter(reference_date.year)
        + month_to_day(date.year, date.month)
        - month_to_day(reference_date.year, reference_date.month)
        + date.day
        - reference_date.day
    )
    chat_ended_days.append(days_passed)

# Add the list of elapsed days as a new column in data
data['chat_ended_days'] = chat_ended_days
```

```python
#Pick the day
from datetime import datetime

chats_week_day = []

for fecha_str in data['chat_started']:
    fecha = datetime.strptime(fecha_str, '%d/%m/%Y')  # Convert the string to a datetime object
    day_number = fecha.strftime('%w')  # Retrieve the day number
    chats_week_day.append(int(day_number))

data['chats_week_day'] = chats_week_day
```

```python
from datetime import datetime

chats_end_week_day = []

for fecha_str in data['chat_ended']:
    fecha = datetime.strptime(fecha_str, '%d/%m/%Y')  # Convert the string to a datetime object
    day_number = fecha.strftime('%w')  # Retrieve the day number
    chats_end_week_day.append(int(day_number))

data['chats_end_week_day'] = chats_end_week_day
```

```python
# Reemplace the week day correctly

def week_clean(x):
    if x in ['Thrusday']:
        return 'Thursday'
    elif x in ['Wensday']:
        return 'Wednesday'
    else:
        return x

data['Active_day'] = list(map(week_clean, data['Active_day']))
data['Active_day'].unique() 
```

```python
# Active_day Change for a number
import pandas as pd

# Definir el diccionario de mapeo de días
day_mapping = {
    'Monday': 1,
    'Tuesday': 2,
    'Wednesday': 3,
    'Thursday': 4,
    'Friday': 5,
    'Saturday': 6,
    'Sunday': 0
}

# Crear una nueva columna con los valores numéricos de los días
data['chat_started_day_num'] = data['Active_day'].apply(lambda x: day_mapping.get(x))
```

```python
#Dropping client_id

data = data.drop(['client_id'], axis=1) 
```

```python
# Replacing NaN value, I saw that my mode is nan so, I'll replacing them for the one that have more sells.

import pandas as pd

def product_fill(x):
    if pd.isna(x):
        return 'Rubí'
    else:
        return x

data['Product'] = data['Product'].apply(product_fill)
```

```python
import pandas as pd

def Product_price_fill(x):
    if pd.isna(x):
        return 1900.0  
    else:
        return x

data['Product_price'] = data['Product_price'].apply(Product_price_fill)
```

```python
import pandas as pd

def Revenue_fill(x):
    if pd.isna(x):
        return 1900.0  
    else:
        return x

data['Revenue'] = data['Revenue'].apply(Revenue_fill)
```
