```python
import seaborn as sns
import matplotlib.pyplot as plt

positive_count = data['positive_comments'].sum()
negative_count = data['negative_comment'].sum()

count_data = pd.DataFrame({'Type': ['Positive Comments', 'Negative Comments'],
                           'Count': [positive_count, negative_count]})

sns.barplot(x='Type', y='Count', data=count_data, palette='viridis')
plt.xlabel('Comment Type')
plt.ylabel('Count')
plt.title('Comments Count')
plt.show()


```

```python
posts_reach_sum = data['posts_reach'].sum()
impressions_sum = data['impressions'].sum()
likes_sum = data['likes'].sum()
stories_reach_sum = data['stories_reach'].sum()

totals_data = pd.DataFrame({'Metric': ['Posts Reach', 'Impressions', 'Likes', 'stories_reach'],
                            'Total': [posts_reach_sum, impressions_sum, likes_sum, stories_reach_sum]})

sns.barplot(x='Metric', y='Total', data=totals_data, palette='viridis')
plt.xlabel('Metric')
plt.ylabel('Total')
plt.title('Instagram Metrics')
plt.show()
```
