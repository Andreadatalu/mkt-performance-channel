# Hypothesis 1:
    
    Looking in the whatsapp contacts, we saw a active day that means the day with the mayority of the contact. The team belive that the active day is the last day that the company keep the contact with the client.
    Our hypothesis is:
        
   * H0: chats_end_week_day != chats_week_day
   * H1: chats_end_week_day = chats_week_day
   
H0: The "chats_end_week_day" is not the most frequently mentioned active day compared to "chats_week_day".
H1: The "chats_end_week_day" is the most frequently mentioned active day compared to "chats_week_day".

```python
import pandas as pd
import numpy as np

file_path = 'C:\\Users\\Andrea\\Documents\\DATA\\Mid-bootcamp project\\Datasets\\LIA _Whatsapp _statistics_clean.csv'
data = pd.read_csv(file_path)
data
```

```python
day_wh = data[['chats_end_week_day', 'chats_week_day']]
day_wh
```

```python
day_wh.dtypes
```

```python
from scipy.stats import chi2_contingency

data_crosstab = pd.crosstab(data['chats_end_week_day'], data['chats_week_day'])

chi2, p_value, _, _ = chi2_contingency(data_crosstab)

print('Chi-square statistic:', chi2)
print('P-value:', p_value)
```

# Hipothesis 2: 

Lia wants to know if the clients with less than 30 years old are more susceptible to paying less for the product.

H0: There is not a significant difference in final prices between people under 30 years of age and people 30 years of age or older.

H1: There is a significant difference in final prices between people under 30 years of age and people 30 years of age or older.

```python
file_path = 'C:\\Users\\Andrea\\Documents\\DATA\\Mid-bootcamp project\\Datasets\\LIA_customer_clean.csv'
data = pd.read_csv(file_path)
data
```

```python
from scipy import stats

clients_1 = data[data['age'] < 30]['final_price']
clients_2 = data[data['age'] >= 30]['final_price']

statistic, pvalue = stats.ttest_ind(clients_1, clients_2)

print('Stat_Score =', statistic)
print('Pvalue =', pvalue)
```
