# What are my cleanning actions

Column | Action
-------|-------
Name | Drop column (high unique values)
Gender | Remplace for F & M
Promotion | Change the % for a value ?
Purchase_day | Split by Year, month and day
Purchase_day | Drop the column
Profit | Change the comma for dot
Columns | Standardize column names
----| -----

```python
#Checking Name values, I found incorrect name, I'll remplace them with an ID.
data = data.drop(['Name'], axis=1) 
data
```

```python
#Gender cleanning

def clean(x):
    if x in ['male']:
        return 'M'
    else:
        return 'F'

data['Gender'] = list(map(clean, data['Gender']))
data['Gender'].unique()  
```

```python
#Transforming the % value into 

def percentage_to_decimal(Promotion):
    if isinstance(Promotion, str) and Promotion.endswith('%'):
        Promotion = Promotion[:-1]  # Eliminar el símbolo de porcentaje
        Promotion_decimal = float(Promotion) / 100
        return Promotion_decimal
    else:
        return Promotion
    

data['Promotion'] = data['Promotion'].apply(percentage_to_decimal)
data['Promotion'].unique()
```

```python
import datetime

data['Purchase_day'] = pd.to_datetime(data['Purchase_day'])
data['PD_Day'] = data['Purchase_day'].dt.day
data['PD_Month'] = data['Purchase_day'].dt.month
data['PD_Year'] = data['Purchase_day'].dt.year
```

```python
#Dropping Purchase_day

data = data.drop(['Purchase_day'], axis=1) 
```

```python
# profit_dot
def profit_dot(value):
    return float(value.replace(',', '.'))

data['Profit'] = data['Profit'].apply(profit_dot)
```

```python
#Standardize column names
Data_cols = data.columns.tolist()


New_cols = [column.lower().replace(' ', '_') for column in Data_cols]


data.columns = New_cols


print(data.columns)
```

```python
#New dataset

data
```

```python
#save my new database

data.to_csv('LIA_customer_clean.csv', index=False)
```
