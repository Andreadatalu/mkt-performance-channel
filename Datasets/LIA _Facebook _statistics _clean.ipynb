# What are my cleanning actions

Column | Action
-------|-------
Date | Split by Year, month and day
Date | Drop the column
Impressions  | Drop nulls values & Convert into Int
Clicks  | Drop nulls values & Convert into Int
Reactions |Fill the #NUM value for de mode
Columns | Standardize column names
----| -----

```python
# Split date

import datetime

data['Date'] = pd.to_datetime(data['Date'])
data['Day'] = data['Date'].dt.day
data['Week'] = data['Date'].dt.week
data['Month'] = data['Date'].dt.month
```

```python
# Drop date

data = data.drop(['Date'], axis=1) 
```

```python
# Drop nulls values on Impressions column
data.dropna(subset=['Impressions'], inplace=True)
```

```python
# Drop nulls values on Clicks column
data.dropna(subset=['Clicks'], inplace=True)
```

```python
#Confirming that is not more nulls values
data.isna().sum()
```

```python
# Convert into Int the Impressions and Clicks columns

data['Impressions'] = data['Impressions'].astype(int)
data['Clicks'] = data['Clicks'].astype(int)
```

```python
#Check is Impressions and Clicks columns are int
data.dtypes
```

```python
# Review Reactions, so, we have a NUM value that doesn't match with the rest. I'll fill them with the mode.

data['Reactions'].value_counts()
```

```python
import statistics as stats
reactions_mode = stats.mode(data['Reactions'])
reactions_mode
```

```python
#Reemplacing #¡NÚM! value for Like (It's the mode)

def reactions_mode(x):
    if x == '#¡NÚM!' :
        return 'Like'
    else:
        return x

data['Reactions'] = data['Reactions'].apply(reactions_mode)
data
```

```python
data['Reactions'].value_counts()
```

```python
#Standardize column names
Data_cols = data.columns.tolist()
New_cols = [column.lower().replace(' ', '_') for column in Data_cols]
data.columns = New_cols

print(data.columns)
```
