# What are my cleanning actions

Column | Action
-------|-------
Date | Split by Year, month and day
Date | Drop the column
Avg_time | Change the value for seconds
rebot_rate  | Change the , in the %
Columns | Standardize column names
----| -----

```python
# Split date

import datetime

data['Date'] = pd.to_datetime(data['Date'])
data['Day'] = data['Date'].dt.day
data['Week'] = data['Date'].dt.week
data['Month'] = data['Date'].dt.month
```

```python
# Drop date

data = data.drop(['Date'], axis=1) 
```

```python
# Change the Avg_time values for seconds
def convert_time_to_seconds(Avg_time):
    hours, minutes, seconds = Avg_time.split(':')
    total_seconds = int(hours) * 3600 + int(minutes) * 60 + int(seconds)
    return total_seconds

data['Avg_time'] = data['Avg_time'].apply(convert_time_to_seconds)
data
```

```python
# Change the , in the %

def rebot_rate_dot(rebot_rate):
    return float(rebot_rate.replace(',', '.'))


data['rebot_rate'] = data['rebot_rate'].apply(rebot_rate_dot)
data['rebot_rate'].unique()
```

```python
#Standardize column names
Data_cols = data.columns.tolist()
New_cols = [column.lower().replace(' ', '_') for column in Data_cols]
data.columns = New_cols

print(data.columns)
```

```python
#New dataset

data
```

```python
#save my new database

data.to_csv('LIA_website_clean.csv', index=False)
```
