# What are my cleanning actions

Column | Action
-------|-------
Results | Convert into Int
Cost per results |Change comma per dot
Clicks |  Convert into Int
Investment | Change comma per dot
Ending | Split by Year, month and day
Columns | Standardize column names
----| -----

```python
# Convert results into INT
data['Results'] = data['Results'].astype(int)
data
```

```python
# Cost_per_results_dot
def Cost_per_results_dot(value):
    return float(value.replace(',', '.'))

data['Cost per results'] = data['Cost per results'].apply(Cost_per_results_dot)
```

```python
# Investment ($) remplance comma per dot and drop colum
def investment_dot(value):
    return float(value.replace(',', '.'))

data['Investment'] = data['Investment'].apply(investment_dot)
data
```

```python
#Clicks dot -  Convert into Int
data['Clicks'] = data['Clicks'].astype(int)
```

```python
# Split ending

import datetime
import pandas as pd

data['Ending'] = pd.to_datetime(data['Ending'], errors='coerce')
data['End_Day'] = data['Ending'].dt.day
data['End_Week'] = data['Ending'].dt.week
data['End_Month'] = data['Ending'].dt.month
```

```python
# Drop date

data = data.drop(['Ending'], axis=1) 
```

```python
#Standardize column names
Data_cols = data.columns.tolist()
New_cols = [column.lower().replace(' ', '_') for column in Data_cols]
data.columns = New_cols

print(data.columns)
```
