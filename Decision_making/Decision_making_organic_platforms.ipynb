```python
data_per = pd.DataFrame({
    'Platform': ['email','f', 'i', 'wh'],
    'Total_dm': [1018, 4874, 4856, 169],
         'Total_sales':[96, 40, 45,36],
        'Convertion_rate': [9.44, 0.82, 0.93,21.30],
        'Profit': [170268.0, 69249.6, 77666.4, 63067.2],
        'Total_profit': [380251.20,380251.20,380251.20,380251.20],
        'Porc_Profit':[44.78, 18.21,20.45,16.59],
        'Proft_per_sale': [1773.63, 1731.24, 1725.92,1751.86]
})


data_per
```

```python
columns_to_include = ['Total_dm','Total_sales','Convertion_rate', 'Porc_Profit', 'Proft_per_sale']

filtered_data = data_per.filter(columns_to_include)

platform_compar = pd.pivot_table(filtered_data, index=data_per['Platform'], aggfunc='sum')
platform_compar = platform_compar[columns_to_include]
platform_compar = platform_compar.reindex(['email','f', 'i', 'wh'])

platform_compar
```
