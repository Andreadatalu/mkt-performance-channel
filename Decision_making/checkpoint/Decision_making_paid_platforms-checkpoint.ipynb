```python
data_per = pd.DataFrame({
    'Platform': ['f', 'g', 'i'],
    'total_impressions': [4233999.00, 9948189.00, 4215066.00],
    'total_clicks': [370227.00, 1752919.00, 336223.00],
    'ctr': [8.74, 17.62, 7.98],
    'total_interactions': [963191.00, 1024895.00, 818786.00],
    'total_investment': [72434.00, 149355.00, 86429.00],
    'Cost_impression': [0.02, 0.02, 0.02],
    'Cost_clicks': [0.20, 0.09, 0.26],
    'Cost_interactions': [0.08, 0.15, 0.16],
    'Profit_impression':[0.09, 0.03, 0.10],
    'Profit_clicks':[1.04, 0.19, 1.31],
    'Profit_interactions':[0.40, 0.32, 0.54],
    'ROI': [433.98, 121.86, 408.70]
})
```

```python
columns_to_include = ['Cost_impression', 'Profit_impression', 'Cost_clicks', 'Profit_clicks', 'Cost_interactions', 'Profit_interactions', 'total_investment', 'ROI']

filtered_data = data_per.filter(columns_to_include)

platform_compar = pd.pivot_table(filtered_data, index=data_per['Platform'], aggfunc='sum')
platform_compar = platform_compar[columns_to_include]
platform_compar = platform_compar.reindex(['f', 'g', 'i'])

# Calcular las columnas de las restas
platform_compar['Value_impression'] = platform_compar['Profit_impression'] - platform_compar['Cost_impression']
platform_compar['Value_clicks'] = platform_compar['Profit_clicks'] - platform_compar['Cost_clicks']
platform_compar['Value_interactions'] = platform_compar['Profit_interactions'] - platform_compar['Cost_interactions']

# Reordenar las columnas
new_columns_order = ['Cost_impression', 'Profit_impression','Value_impression', 'Cost_clicks', 'Profit_clicks','Value_clicks', 'Cost_interactions', 'Profit_interactions','Value_interactions', 'total_investment', 'ROI']
platform_compar = platform_compar.reindex(new_columns_order, axis=1)

platform_compar.T
```

```python
total_investment_total = data_per['total_investment'].sum()
data_per['investment_difference'] = (data_per['total_investment'] / total_investment_total) * 100 

data_per.T
```
